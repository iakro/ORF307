```python
import pandas as pd
```

```python
# all data is courtesy of investing.com
```

```python
tickers = [
    'AMD',
    'AMZN',
    'BAC',
    'GOOG',
    'GS',
    'K',
    'KO',
    'META',
    'NKE',
    'SHEL',
    'SPY',
]
```

```python
out_df = pd.DataFrame()
```

```python
for ticker in tickers:
    fname = f'data/{ticker}.csv'
    curr_df = pd.read_csv(fname)[::-1]
#     curr_df = curr_df.reindex(index=curr_df.index[::-1])
    curr_change = curr_df['Change %'].str.rstrip('%').astype(float)
    print(curr_change)
    out_df[ticker] = curr_change
```

```python
out_df
```

```python
out_df.to_csv('returns.csv', index=False)
```
